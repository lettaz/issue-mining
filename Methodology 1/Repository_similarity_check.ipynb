```python
import pandas as pd

# Read the complete_analysis.csv
df_complete_analysis = pd.read_csv('complete_analysis.csv')

# Get the first three and last three rows
top_three = df_complete_analysis.head(3)
bottom_three = df_complete_analysis.tail(3)

# Concatenate the top three and bottom three into a new DataFrame
df_combined = pd.concat([top_three, bottom_three], axis=0)

# Read the seed_dataset.csv
df_seed = pd.read_csv('../seed_dataset.csv')

# Initialize an empty DataFrame to store the new dataset
df_repository_similarity_check = pd.DataFrame()

# Loop through the rows in the combined DataFrame
for i, row in df_combined.iterrows():
    # Search for the repository in the seed dataset
    repo_data = df_seed[df_seed['name'] == row['Repository']]
    # If the repository was found in the seed dataset, append the data to the new DataFrame
    if not repo_data.empty:
        repo_data['Issue ID'] = row['Issue ID']
        repo_data['Issue Type'] = row['Issue Type']
        repo_data['Creation Date'] = row['Creation Date']
        repo_data['Resolution Date'] = row['Resolution Date']
        repo_data['Status'] = row['Status']
        repo_data['Status Change Date'] = row['Status Change Date']
        repo_data['Number of Comments'] = row['Number of Comments']
        repo_data['Labels'] = row['Labels']
        repo_data['Completion Time'] = row['Completion Time']
        df_repository_similarity_check = df_repository_similarity_check.append(repo_data)

# Save the new DataFrame to repository_similarity_check.csv
df_repository_similarity_check.to_csv('repository_similarity_check.csv', index=False)
```
